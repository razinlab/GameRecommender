```python
import pandas as pd
```

```python
df = pd.read_csv('GameData.csv')
```

```python
df
```

```python
def row_to_string(row):
    game_string = f'''
    Game: {row['name']},
    Summary: {row['summary']},
    Story: {row['storyline']},
    Rating: {row['aggregated_rating']},
    Genres: {row['genres']},
    Platforms: {row['platforms']},
    Themes: {row['themes']},
    DLCs: {row['dlcs']},
    DLC Count: {row['dlc_count']},
    Expansion Count: {row['expansion_count']},
    Game Engine: {row['game_engines']},
    Involved Companies: {row['involved_companies']},
    Keywords: {row['keywords']},
    Similar Games: {row['similar_games']},
    Player Perspective: {row['player_perspectives']},
    Cover: {row['cover_url']},
    Steam: {row['steam_url']},
    Gog: {row['gog_url']},
    Playstation: {row['playstation_url']},
    Epic Games: {row['epic_url']},
    Nintendo: {row['nintendo_url']},
    Official: {row['official_url']},
    Release Date UNIX: {row['release_date']},
    Release Date: {row['release_date_human']},
    '''
    return game_string
```

```python
df['row_text'] = df.apply(row_to_string, axis=1)
```

```python
print(df['row_text'].values[0])
```

```python
import faiss
import requests
import numpy as np
```

```python
DIM = 1024
INDEX = faiss.IndexFlatL2(DIM)

x = np.zeros((len(df['row_text']), DIM), dtype='float32')
```

```python
for i, text in enumerate(df['row_text']):
    if i % 500 == 0:
        print(f'Processed {i} instances')

    res = requests.post('http://localhost:11434/api/embeddings',
                        json={
                            'model':'snowflake-arctic-embed2',
                            'prompt': text,
                        })

    embedding = res.json()['embedding']

    x[i] = np.array(embedding)

INDEX.add(x)
```

```python
faiss.write_index(INDEX, 'index')
```
